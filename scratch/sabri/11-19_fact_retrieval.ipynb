```python
import os

import pandas as pd
from tqdm import tqdm
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from zoology.analysis.utils import fetch_wandb_runs
```

```python
df = fetch_wandb_runs(
    sweep_id=["fact_retrievalb1770b"], 
    project_name="zoology"
)
df = df.groupby(["model.d_model", "data.0.builder.kwargs.n_subjects"])["valid/accuracy"].max().reset_index()
```

```python
# Assuming df is your DataFrame
pivot_table = df.pivot_table(values='valid/accuracy', index='data.0.builder.kwargs.n_subjects', columns='model.d_model')

plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title('Validation Accuracy Heatmap')
plt.show()
```
